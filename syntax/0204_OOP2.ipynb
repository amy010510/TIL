### super() ì •ë¦¬
- MRO ì´í•´ í•„ìˆ˜!
- ClassName.__mro__ë˜ëŠ” **ClassName.mor()**ë¥¼ í™•ì¸í•´ MRO ìˆœì„œí™•ì¸ê°€ëŠ¥

```python
class A:
    def __init__(self):
        print('A Constructor')


class B(A):
    def __init__(self):
        super().__init__()
        print('B Constructor')


class C(A):
    def __init__(self):
        super().__init__()
        print('C Constructor')


class D(B, C):
    def __init__(self):
        super().__init__()
        print('D Constructor')


# [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
print(D.mro())

# (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
print(D.__mro__)
```
